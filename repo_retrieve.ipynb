```python
def search_github_repositories(query, sort="stars", order="desc", per_page=10, page=1):
    """
    Search for public repositories on GitHub using the REST API.
    
    Args:
        query (str): The search query (e.g., 'pytorch in:readme').
        sort (str): The sorting field ('stars', 'forks', or 'updated').
        order (str): The sorting order ('asc' or 'desc').
        per_page (int): Number of results per page (max: 100).
        page (int): Page number to retrieve.
        
    Returns:
        list: A list of repository details or an error message.
    """
    headers = {
        "Authorization": f"Bearer {GITHUB_TOKEN}",
        "Accept": "application/vnd.github.v3+json",
    }
    params = {
        "q": query,
        "sort": sort,
        "order": order,
        "per_page": per_page,
        "page": page,
    }

    response = requests.get(API_URL, headers=headers, params=params)

    if response.status_code == 200:
        return response.json().get("items", [])
    else:
        print(f"Error: {response.status_code}, {response.json()}")
        return []

# Example Usage
if __name__ == "__main__":
    # Search for repositories mentioning "pytorch" in the README
    repositories = search_github_repositories("pytorch in:readme in:description topic:pytorch")

    print(f"Found {len(repositories)} repositories:\n")
    for repo in repositories:
        print(f"Name: {repo['name']}")
        print(f"Owner: {repo['owner']['login']}")
        print(f"Stars: {repo['stargazers_count']}")
        print(f"URL: {repo['html_url']}")
        print(f"Description: {repo['description']}\n")
```

```python
def get_total_repository_count(query):
    """
    Fetch the total count of repositories matching the query on GitHub.
    
    Args:
        query (str): The search query (e.g., 'pytorch in:readme in:description topic:pytorch stars:>1000').
    
    Returns:
        int: Total count of matching repositories or -1 if an error occurs.
    """
    headers = {
        "Authorization": f"Bearer {GITHUB_TOKEN}",
        "Accept": "application/vnd.github.v3+json",
    }
    params = {
        "q": query,
    }

    response = requests.get(API_URL, headers=headers, params=params)

    if response.status_code == 200:
        return response.json().get("total_count", 0)
    else:
        print(f"Error: {response.status_code}, {response.json()}")
        return -1

# Example Usage
if __name__ == "__main__":
    # Base query for PyTorch repositories
    base_query = "topic:pytorch"

    # Get counts for different star thresholds
    thresholds = [1000, 500, 100]
    for threshold in thresholds:
        query_stars = f"{base_query} stars:>{threshold} "
        total_count = get_total_repository_count(query_stars)
        if total_count != -1:
            print(f"Total repositories with stars > {threshold}: {total_count}")
        else:
            print(f"Failed to fetch the count for stars > {threshold}.")


        query_forks = f"{base_query} forks:>={threshold} "
        count = get_total_repository_count(query_forks)
        if count != -1:
            print(f"Total repositories with forks > {threshold}: {count} \n")
        else:
            print(f"Failed to fetch count for  with stars: {threshold}")
```
